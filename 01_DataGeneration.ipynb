```python
from faker import Faker
import pandas as pd
import random
from datetime import datetime, timedelta

fake = Faker()
NUM_CUSTOMERS = 500
NUM_PURCHASES = 5000
segments = ["Loyal", "Occasional", "New"]
segment_weights = [0.3, 0.5, 0.2]

customers = []
for i in range(1, NUM_CUSTOMERS + 1):
    segment = random.choices(segments, weights=segment_weights, k=1)[0]
    customers.append({
        "CustomerID": f"C{i:04}",
        "Name": fake.name(),
        "Region": random.choice(["North", "South", "East", "West"]),
        "JoinDate": fake.date_between(start_date="-2y", end_date="today"),
        "Segment": segment
    })

purchases = []
for i in range(1, NUM_PURCHASES + 1):
    cust = random.choice(customers)
    segment = cust["Segment"]
    if segment == "Loyal":
        amount = round(random.uniform(200, 1200), 2)
        days_back = random.randint(1, 90)
    elif segment == "Occasional":
        amount = round(random.uniform(50, 600), 2)
        days_back = random.randint(10, 180)
    else:
        amount = round(random.uniform(20, 300), 2)
        days_back = random.randint(30, 365)

    purchase_date = datetime.today() - timedelta(days=days_back)
    purchases.append({
        "PurchaseID": f"P{i:05}",
        "CustomerID": cust["CustomerID"],
        "Date": purchase_date.strftime("%Y-%m-%d"),
        "Amount": amount
    })

df_customers = pd.DataFrame(customers)
df_purchases = pd.DataFrame(purchases)

# Save to Lakehouse folders
lakehouse_path = "/lakehouse/default/Files/raw/"
df_customers.to_csv(f"{lakehouse_path}customers.csv", index=False)
df_purchases.to_csv(f"{lakehouse_path}purchases.csv", index=False)

print("✅ Data generation complete and saved to Lakehouse!")
```
